```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


plt.style.use("ggplot")
```

# Задание 1

```python
n = 9


x = np.arange(-1.0,1.0, 0.01)
y = np.arange(-1.0,1.0, 0.01)

x, y = np.meshgrid(x,y)

f = lambda x, y: (np.sin(y) + x ** 2) * np.exp(x ** n + y) + y ** n
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 5))

c = ax[0].contour(x, y, f(x, y), n, colors='red')
cs = ax[0].contourf(x, y, f(x, y), n)
ax[0].clabel(c, inline=1, fontsize=10, colors='red')

c = ax[1].contour(x, y, f(x, y), n, colors='red')
ax[1].clabel(c, inline=1, fontsize=10, colors='red')
```

# Задание 2

```python
import matplotlib.scale as sc
fig, ax = plt.subplots(2, 2, figsize=(20, 10))

f = lambda x: np.sin(x) * np.cos(2 * x)
x = np.arange(0, 15, 0.01)
ax[0][0].plot(x, f(x))
ax[0][0].set_xlabel('x')
ax[0][0].set_ylabel('f(x)')


ax[0][1].plot(x * 5, f(x * 5))
ax[0][1].set_xlabel('x scaled by 5')
ax[0][1].set_ylabel('f(x) scaled by 5')


f = lambda x: -n * 2/x + np.cos(x) + np.sin(x ** 3)
x = np.arange(-17, 4, 0.01)
ax[1][0].plot(x, f(x))
ax[1][0].set_xlabel('x')
ax[1][0].set_ylabel('f(x)')


ax[1][1].plot(x * 5, f(x * 5))
ax[1][1].set_xlabel('x scaled by 5')
ax[1][1].set_ylabel('f(x) scaled by 5')
```
