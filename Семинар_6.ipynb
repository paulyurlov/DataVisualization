# Задание 2

```python
df = pd.read_csv('Data/iris.csv')
df.head()
```

```python
_, ax = plt.subplots(1, 2, figsize=(20,5))
ax[0].scatter(df['petal.length'], df['petal.width'])
ax[0].set_title("Соотношение размеров лепестков")
ax[0].set_xlabel('Длинна')
ax[0].set_ylabel('Ширина')
ax[1].scatter(df['sepal.length'], df['sepal.width'])
ax[1].set_title("Соотношение размеров чашечек")
ax[1].set_xlabel('Длинна')
ax[1].set_ylabel('Ширина')
```

```python
_, ax = plt.subplots(1, 2, figsize=(20,5))
sns.scatterplot(data=df, x='petal.length', y='petal.width', hue='variety', ax=ax[0])
# ax[0].scatter(df['petal.length'], df['petal.width'], label=df['variety'])
ax[0].set_title("Соотношение размеров лепестков")
ax[0].set_xlabel('Длинна')
ax[0].set_ylabel('Ширина')
# ax[1].scatter(df['sepal.length'], df['sepal.width'])
sns.scatterplot(data=df, x='sepal.length', y='sepal.width', hue='variety', ax=ax[1])
ax[1].set_title("Соотношение размеров чашечек")
ax[1].set_xlabel('Длинна')
ax[1].set_ylabel('Ширина')
```

```python
_, ax = plt.subplots(2, 2, figsize=(20,10))
sns.histplot(data=df, x='petal.length', hue='variety', ax=ax[0][0])
ax[0][0].set_title("Распределение длинны лепестка")
ax[0][0].set_xlabel('Длинна')

sns.histplot(data=df, x='petal.width', hue='variety', ax=ax[0][1])
ax[0][1].set_title("Распределение ширины лепестка")
ax[0][1].set_xlabel('Ширина')
sns.histplot(data=df, x='petal.length', hue='variety', ax=ax[1][0])
ax[1][0].set_title("Распределение длинны чашечки")
ax[1][0].set_xlabel('Длинна')
sns.histplot(data=df, x='petal.width', hue='variety', ax=ax[1][1])
ax[1][1].set_title("Распределение ширины чашечки")
ax[1][1].set_xlabel('Ширина')
```

# Задание 3

```python
df = pd.read_csv('Data/tips.csv')
df.head()
```

```python
fig = sns.scatterplot(data=df, x='total_bill', y='tip')
_ = fig.set(title='Связь между чаевыми и общим потреблением', xlabel='Счет', ylabel='Чаевые')
```

```python
# plt.bar(data=df.groupby('day').sum(), x='tip')
sns.barplot(data=df, x="day", y="tip", estimator=np.sum,)
```

```python
_, ax = plt.subplots(1, 2, figsize=(20,5))
sns.histplot(data=df, x='tip', hue='sex', ax=ax[0])

ax[0].set_title('Влияние пола на чаевые')
ax[0].set_xlabel('Чевые')
ax[0].set_ylabel('Число')

sns.histplot(data=df, x='tip', hue='smoker', ax=ax[1])
ax[0].set_title('Влияние курения на чаевые')
ax[1].set_xlabel('Чевые')
ax[1].set_ylabel('Число')
```
