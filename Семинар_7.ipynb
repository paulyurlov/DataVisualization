### Пункт 5 - 6

```python
df = pd.read_csv('Data/function.csv', sep=';')
df = df.apply(lambda x: x.str.replace(',','.')).copy()
df = df.astype('float').copy()
print(df.shape)
df.head()
```

```python
_, ax = plt.subplots(2, 2, figsize=(20, 15))

ax[0][0].plot(df.X, df.Y)
ax[0][0].set_title('Линейный масштаб')
ax[0][0].set_xlabel('x')
ax[0][0].set_ylabel('y')

ax[0][1].plot(df.X, df.Y)
ax[0][1].set_title('Логарифмический по оси y')
ax[0][1].set_yscale('log')
ax[0][1].set_xlabel('x')
ax[0][1].set_ylabel('log(y)')

ax[1][0].plot(df.X, df.Y)
ax[1][0].set_title('Логарифмический по оси x')
ax[1][0].set_xscale('log')
ax[1][0].set_xlabel('log(x)')
ax[1][0].set_ylabel('y')

ax[1][1].plot(df.X, df.Y)
ax[1][1].set_title('Наименьшее значение - -α,  наибольшее - + α. В середине – 0')
ax[1][1].set_xlabel('x')
ax[1][1].set_ylabel('y')
_= ax[1][1].set_xticks([df.X.min() , 0 , df.X.max()], ['-a', '0.0', '+a'])
```
